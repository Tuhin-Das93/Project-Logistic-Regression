```python
loan.info()
```

### Outlier Treatment

```python
cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',]

Ql = loan[cols].quantile(0.05)
Qh = loan[cols].quantile(0.95)
IQR = Qh - Ql

loan = loan[~((loan[cols] < (Ql - 1.5 * IQR)) |(loan[cols] > (Qh + 1.5 * IQR))).any(axis=1)]
```

```python
num_loan = loan[['ApplicantIncome','CoapplicantIncome','LoanAmount']]
num_loan.describe(percentiles=[.25, .5, .75, .90, .95])
```
